## Aufgabe 1: Regression

Erklären Sie kurz (mit wenigen Sätzen) die folgenden Begriffe:

* Methode der kleinsten Quadrate (Least squares)
* Gradientenabstieg (Gradien Descent)
* Normal Equation
* Kostenfunktion
* Gradient Check
* Multiple lineare Regression (Multivariate Linear Regression)


**Lösung:**

*Methode der kleinsten Quadrate (Least squares)*
- Algorithmus zum Lösen von Regressionsproblemen
- Minimiert die Summe der Quadrate der Abweichungen von Vorhersage und tatsächlichem Wert der Zielvariablen

*Gradientenabstieg (Gradien Descent)*
- iterativer Optimierungsalgorithmus zum Finden von lokalen Minima von differenzierbaren Funktionen
- Parameter werden in entgegengesetzter Richtung des Gradienten aktualisiert (steilster Abstieg)

*Normal Equation*
- berechnet Regressions-Koeffizienten analytisch für die Multivariate Lineare Regression
- Im Gegensatz zu Gradient Descent benötigt der Algorithmus nur einen Schritt

*Kostenfunktion*
- bewertet die Abweichung der Vorhersage vom tatsächlichen Wert einer Zielvariablen

*Gradient Check*
- Berechnung des numerischen Gradienten zur Überprüfung der Implementierung des analytischen Gradienten
- Numerischer Gradient ist einfach zu implementieren, aber sehr rechenaufwändig
- Analytischer Gradient ist schnell zu berechnen, aber fehleranfälliger in der Implementierung

*Multiple lineare Regression (Multivariate Linear Regression)*
- versucht eine beobachtete abhängige Variable durch mehrere unabhängige Variablen zu erklären
- im Gegensatz zur einfachen linearen Regression werden mehrere Variablen betrachtet

b) Das Gradientenabstiegsverfahren mit Momentum ist eine Modifikation des Gradientenabstiegsverfahrens, 
bei dem statt der direkten Aktualisierung der Parameter zunächst ein Momentums-Vektor $\Delta$ aktualisiert wird, 
der nicht nur vom Gradienten, sondern auch dem Momentum der letzten Iteration abhängt. 
Dabei wird über den Parameter $\mu$ gesteuert, wie stark das Momentum gewichtet wird.

$\Delta := - \alpha \frac{\partial}{\partial u} f(u)+ \mu \Delta$

Der Momentums-Vektor wird nun verwendet, um $u$ zu aktualisieren. Die Update-Regel ändert sich damit zu

$u := u + \Delta $

d) Implementieren sie das Gradientenabstiegsverfahren mit Momentum zur Bestimmung des Minimums der Funktion $f$ in Python. 

*Hinweis: Verwenden Sie in dieser Aufgabe keine Bibliotheken außer Numpy.*

```python
f = lambda u: ((u-10)**4)/4 + (u-10)**3 + 50
fd = lambda u: (u-10)**3 + 3*(u-10)**2

def gradUpdate(fDer, u, delta, eps, mu):
    g = fDer(u)
    delta = -eps * g + delta * mu
    u = u + delta
    return u, g, delta

def gradDesc(fDer, n, startU, startDelta, eps, mu):
    
    us = [startU]
    grads = [0]
    deltas = [0]
    
    for i in range(n):    
        u, grad, delta = gradUpdate(fDer, us[i], deltas[i], eps, mu)
        us.append(u)
        grads.append(grad)
        deltas.append(delta)
        
    return us, grads, deltas
```

```python
# Check Values from task a)

us_a, grads_a, _ = gradDesc(fd, 5, 13, 0, 0.05, 0)

for i in range(len(us_a)):
    print('iteration {:2d}:   grad={:8.3f}  =>  u={:7.3f}'.format(i, grads_a[i], us_a[i]))
```

```python
# Check Values from task b)

us_b, grads_b, deltas_b = gradDesc(fd, 5, 13, 0, 0.05, 0.5)

for i in range(len(us_b)):
    print('iteration {:2d}:   grad={:8.3f}  delta={:7.3f}  =>  u={:7.3f}'.format(i, grads_b[i], deltas_b[i], us_b[i]))
```
