# To Compare our results

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv ('../dataset/predictions_compare.csv', index_col='Id')

# Extract data
zoe_results = df['Zoe']
hiro_results = df['Hiro']

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(zoe_results, hiro_results, color='blue', alpha=0.5)
plt.title('Comparison of Zoe vs Hiro Results')
plt.xlabel('Zoe Results')
plt.ylabel('Hiro Results')
plt.grid(True)
plt.show()
```

```python
# Box plots
plt.figure(figsize=(8, 6))
plt.boxplot([df['Zoe'], df['Hiro']], labels=['Zoe', 'Hiro'])
plt.title('Box Plot of Zoe and Hiro Results')
plt.ylabel('Results')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(df['Zoe'], bins=30, alpha=0.5, label='Zoe Results')
plt.hist(df['Hiro'], bins=30, alpha=0.5, label='Hiro Results')
plt.title('Histogram of Zoe and Hiro Results')
plt.xlabel('Results')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
